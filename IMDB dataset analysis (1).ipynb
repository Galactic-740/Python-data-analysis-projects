# Top 5 director with the highest revenue of each movies 

```python
direcrtors_movie_revenue=new_df.groupby("Director").agg({"Revenue (Millions)":"sum"}).reset_index()
top_5=direcrtors_movie_revenue.sort_values("Revenue (Millions)",ascending=False).nlargest(5,"Revenue (Millions)")
top_5
```

```python
#textprops={"color":"g","fontstyle":"italic","fontsize":12}
#wedgecolor={"linewidth":1,"width":1,"edgecolor":"k"}
plt.pie(top_5["Revenue (Millions)"], labels=top_5["Director"],autopct="%11f%%")
plt.show()
```

Visualize the top 10 genre  which have  highest revenue by years 

We get from this Action,Adventure,Sci-Fi genre movies has higest Revenue

```python
Genre_revenue_yeras=new_df.groupby(["Genre","Year"]).agg({"Revenue (Millions)": "sum"}).reset_index().sort_values("Revenue (Millions)",ascending=False)
Top_10_genre_revenue=Genre_revenue_yeras.nlargest(10,"Revenue (Millions)")
Top_10_genre_revenue
```

```python
plt.figure(figsize=(25,10))
sns.set(style="darkgrid")
ax=sns.lineplot(data=Top_10_genre_revenue,x="Year",y="Revenue (Millions)",hue="Genre", markers=["o","*",">"],
             style="Genre",palette="hot",markersize=30)
ax.set_facecolor("black")
plt.legend(fontsize=25,facecolor="g",labelcolor="w")
plt.title("Top_10 Genre through highest revenue by years",fontsize=30)
plt.xlabel("Year",size=35,)
plt.ylabel("Revenue (Millions)",size=35)
plt.tick_params(axis="both",labelsize=30)
plt.show()
```
